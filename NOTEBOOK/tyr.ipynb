```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
from sklearn.model_selection import GridSearchCV
```

```python
grboost = GradientBoostingClassifier(learning_rate=0.1, max_depth=3, n_estimators=50)
grboost.fit(X_train, y_train)
```

```python
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
results = {'Accuracy': [], 'Precision': [], 'Recall': [], 'F1 Score': [], 'ROC AUC': [], 'Confusion Matrix': []}

for train_index, valid_index in kf.split(X_train, y_train):
    X_trainv, X_valid = X_train.iloc[train_index], X_train.iloc[valid_index]
    y_trainv, y_valid = y_train.iloc[train_index], y_train.iloc[valid_index]

    model = GradientBoostingClassifier()
    model.fit(X_trainv, y_trainv)

    y_pred = model.predict(X_valid)
    y_prob = model.predict_proba(X_valid)[:, 1]

    results['Accuracy'].append(accuracy_score(y_valid, y_pred))
    results['Precision'].append(precision_score(y_valid, y_pred))
    results['Recall'].append(recall_score(y_valid, y_pred))
    results['F1 Score'].append(f1_score(y_valid, y_pred))
    results['ROC AUC'].append(roc_auc_score(y_valid, y_prob))
    results['Confusion Matrix'].append(confusion_matrix(y_valid, y_pred))

final_gb_results = {
    'Accuracy Mean': np.mean(results['Accuracy']),
    'Accuracy Std': np.std(results['Accuracy']),
    'Precision Mean': np.mean(results['Precision']),
    'Precision Std': np.std(results['Precision']),
    'Recall Mean': np.mean(results['Recall']),
    'Recall Std': np.std(results['Recall']),
    'F1 Score Mean': np.mean(results['F1 Score']),
    'F1 Score Std': np.std(results['F1 Score']),
    'ROC AUC Mean': np.mean(results['ROC AUC']),
    'ROC AUC Std': np.std(results['ROC AUC']),
    'Confusion Matrix': str(results['Confusion Matrix'][-1])
}

results_gb_df = pd.DataFrame([final_gb_results])
results_gb_df
```

```python
mean_metrics = [
    final_gb_results['Accuracy Mean'],
    final_gb_results['Precision Mean'],
    final_gb_results['Recall Mean'],
    final_gb_results['F1 Score Mean'],
    final_gb_results['ROC AUC Mean']
]

std_metrics = [
    final_gb_results['Accuracy Std'],
    final_gb_results['Precision Std'],
    final_gb_results['Recall Std'],
    final_gb_results['F1 Score Std'],
    final_gb_results['ROC AUC Std']
]

metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'ROC AUC']
colors = ['skyblue', 'salmon', 'lightgreen', 'orange', 'lightcoral']

plt.figure(figsize=(10, 6))
plt.title('Gradient Boosting Performance', fontsize=20)

plt.bar(metrics, mean_metrics, yerr=std_metrics, capsize=5, color=colors)

plt.yticks(np.arange(0, 1.05, 0.05))
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

plt.ylabel('Scores', fontsize=14)
plt.tight_layout()

plt.show()
```

```python
def optimize_hyperparameters(model, param_grid, X_train, y_train, scoring):
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring=scoring, n_jobs=-1, verbose=1)
    grid_search.fit(X_train, y_train)
    return grid_search.best_estimator_, grid_search.best_params_, grid_search.best_score_
```

```python
param_grid = {
    'n_estimators': [50, 100, 200, 300, 400],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 4, 5, 6]
}

scoring_metrics = ['roc_auc', 'accuracy', 'precision', 'recall', 'f1']

optimized_results = {}

for scoring in scoring_metrics:
    print(f"\n Optimizing GradientBoostingClassifier for scoring: {scoring}")

    best_model, best_params, best_score = optimize_hyperparameters(
        GradientBoostingClassifier(),
        param_grid,
        X_train,
        y_train,
        scoring
    )

    optimized_results[scoring] = {
        'Best Model': best_model,
        'Best Params': best_params,
        f'Best {scoring} Score': best_score
    }

optimized_results
```

```python
pd.set_option('display.max_colwidth', None)

dataframes = {}

for scoring, model_info in optimized_results.items():
    df = pd.DataFrame([{
        'Model': 'GradientBoostingClassifier',
        'Best Params': model_info['Best Params'],
        f'Best {scoring} Score': model_info[f'Best {scoring} Score']
    }])

    dataframes[scoring] = df

for scoring, df in dataframes.items():
    print(f"DataFrame for {scoring} scoring metric:")
    display(df)
    print("\n" + "="*50 + "\n")
```

```python
print("Available scoring metrics:", list(optimized_results.keys()))
```

```python
best_scoring_metric = 'roc_auc' 
```

```python
best_info = optimized_results[best_scoring_metric]
best_params = best_info['Best Params']

print(f"\nUsing model optimized for: {best_scoring_metric}")
print("Best hyperparameters:")
print(best_params)
```

```python
final_model = GradientBoostingClassifier(**best_params)
```

```python
final_model.fit(X_train, y_train)
print("\n Final GradientBoosting model trained on full training set.")
```

### TESTING THE MODEL

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report
)

y_pred = final_model.predict(X_test)

y_prob = final_model.predict_proba(X_test)[:, 1]

test_accuracy = accuracy_score(y_test, y_pred)
test_precision = precision_score(y_test, y_pred)
test_recall = recall_score(y_test, y_pred)
test_f1 = f1_score(y_test, y_pred)
test_roc_auc = roc_auc_score(y_test, y_prob)
test_cm = confusion_matrix(y_test, y_pred)

print("\n Test Evaluation (GradientBoostingClassifier)")
print(f"Accuracy   : {test_accuracy:.3f}")
print(f"Precision  : {test_precision:.3f}")
print(f"Recall     : {test_recall:.3f}")
print(f"F1 Score   : {test_f1:.3f}")
print(f"ROC AUC    : {test_roc_auc:.3f}")
print("\nConfusion Matrix:\n", test_cm)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
